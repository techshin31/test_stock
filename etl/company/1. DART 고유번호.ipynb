```python
from dotenv import load_dotenv
import os 

load_dotenv()
api_key = os.getenv("DART_API_KEY")
```

```python
# 라이브러리 설치
# pip install opendartreader

import OpenDartReader

# 1. OpenDartReader 객체 생성 (본인의 API Key 입력)
dart = OpenDartReader(api_key)
```

```python
import pandas as pd

df_com = pd.read_csv("./data/wics_company_2026.csv")[['CMP_CD', 'CMP_KOR']]
df_com.drop_duplicates(subset=['CMP_CD'], inplace=True)

df_com.shape  
```

```python
df_com['DART_CD'] = df_com['CMP_CD'].apply(lambda x: dart.find_corp_code(x))
```
