```python
import os
import OpenDartReader
from dotenv import load_dotenv

load_dotenv()
api_key = os.getenv("DART_API_KEY")

# 인증키 설정
dart = OpenDartReader(api_key)

# 대상 기업 고유번호 및 연도 설정
target_corp = "00126380"  # 삼성전자
target_year = 2025
```

```python
from dart_api import fetch_business_overview

# 1. 사업의 개요 — Open DART의 dart.report()에는 '사업의개요' 키가 없음.
#    사업보고서 접수번호 → DART 웹 목차(sub_docs) → viewer HTML에서 본문 텍스트 추출.
biz_overview = fetch_business_overview(dart, target_corp, target_year)
```

```python
from dart_api import fetch_segment_sales_tables

# 2. 주요 사업부별 성과 — report() API에 '부문별재무현황' 없음. 사업보고서 내 표(예: 매출 및 수주상황) 파싱.
segment_tables = fetch_segment_sales_tables(dart, target_corp, target_year)
print("\n--- 사업부별 매출 등 (사업보고서 표, 첫 번째 후보) ---")
if segment_tables:
    df_segment = segment_tables[0]
    print(df_segment.to_string())
else:
    df_segment = pd.DataFrame()
    print("부문/매출 표를 찾지 못했습니다. 공시 목차 제목이 다른 기업이면 match 키워드를 조정하세요.")
```

```python
df_segment.drop_duplicates(inplace=True)
df_segment.head() 
```

```python
# 3. 감사의견 — OpenDartReader 키워드는 '회계감사'(회계감사인의 명칭 및 감사의견 API).
#    컬럼: adtor(감사인), adt_opinion(감사의견), adt_reprt_spcmnt_matter, emphs_matter, core_adt_matter 등
df_audit = dart.report(target_corp, "회계감사", target_year)
print("\n--- 감사의견 정보 ---")
if df_audit is not None and not df_audit.empty:
    print(df_audit[["adtor", "adt_opinion"]].drop_duplicates())
else:
    print(df_audit)
```
