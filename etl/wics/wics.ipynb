# wics_major.json

```python
import json 
import pandas as pd 
```

```python
with open('./row/wics_major.json', 'r', encoding='utf-8') as f:
    data = json.load(f)
```

```python
data.keys()
```

```python
df = pd.json_normalize(data, record_path=['items'], meta=['description'])
df.head()
```

```python
df.rename(columns={
        'code': 'wics_major_code', 'name': 'wics_major_name', 'description': 'wics_major_description'
    }, 
    inplace=True)
df.head()
```

```python
df.shape
```

# wics_medium.json

```python
with open('./row/wics_medium.json', 'r', encoding='utf-8') as f:
    data = json.load(f)
```

```python
df_medium = pd.json_normalize(data, record_path=['items'], meta=['description'])
df_medium.head()
```

```python
df_medium.rename(columns={
    'code': 'wics_medium_code', 'name': 'wics_medium_name', 'description': 'wics_medium_description',
    'major_code': 'wics_major_code'
    }, 
    inplace=True)
df_medium.head()
```

```python
df_medium.shape
```

## merge

```python
df = df.merge(df_medium, on='wics_major_code', how='left')
df.head()
```

# wics_minor.json

```python
with open('./row/wics_minor.json', 'r', encoding='utf-8') as f:
    data = json.load(f)

df_minor = pd.json_normalize(data, record_path=['items'], meta=['description'])

df_minor.rename(columns={
    'code': 'wics_minor_code', 'name': 'wics_minor_name', 'description': 'wics_minor_description',
    'medium_code': 'wics_medium_code'
    }, 
    inplace=True)
df_minor.head()
```

```python
df_minor.shape
```

## merge

```python
df = df.merge(df_minor, on='wics_medium_code', how='left')
df.head()
```

```python
df.shape 
```

# wics_detailed.json

```python
with open('./row/wics_detailed.json', 'r', encoding='utf-8') as f:
    data = json.load(f)
df_detailed = pd.json_normalize(data, record_path=['items'])

df_detailed.rename(columns={
    'code': 'wics_detailed_code', 'name': 'wics_detailed_name', 'description': 'wics_detailed_description',
    'minor_code': 'wics_minor_code'
    }, 
    inplace=True)

df_detailed.head()
```

```python
df_detailed.shape
```

## merge

```python
df = df.merge(df_detailed, on='wics_minor_code', how='left')
df.head()
```
