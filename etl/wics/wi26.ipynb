# wi26_major.json

```python
import json 
import pandas as pd 
```

```python
with open('./row/wi26_major.json', 'r', encoding='utf-8') as f:
    data = json.load(f)
```

```python
df = pd.json_normalize(data, record_path=['items'], meta=['description'])
df.rename(columns={
    'code': 'wi26_major_code', 'name': 'wi26_major_name', 'description': 'wi26_major_description'
    }, 
    inplace=True)
df.head()
```

```python
df.shape  
```

# wi26_minor.json

```python
with open('./row/wi26_minor.json', 'r', encoding='utf-8') as f:
    data = json.load(f)
```

```python
df_minor = pd.json_normalize(data, record_path=['items'], meta=['description'])
df_minor.rename(columns={
    'code': 'wi26_minor_code', 'name': 'wi26_minor_name', 'description': 'wi26_minor_description',
    'major_code': 'wi26_major_code'
    }, 
    inplace=True)
df_minor.head()
```

```python
df_minor.shape  
```

## merge

```python
df.columns 
```

```python
df_minor.columns 
```

```python
df = df.merge(df_minor, on='wi26_major_code', how='left')
df.head()
```

```python
df.shape 
```
