# Bollinger Bands (변동성 + 가격 위치)
- 가격이 밴드 상단/하단 어디에 있는지 확인
- 핵심:
    - 상단 터치 → 과열 가능
    - 하단 터치 → 반등 가능
-  왜 중요?
    - 변동성까지 같이 고려 가능
    - 횡보장에서 특히 강함

# RSI (상대 강도 지수)
- 과매수 / 과매도 판단의 대표 지표
- 기준:
    - 70 이상 → 과매수 (매도 고려)
    - 30 이하 → 과매도 (매수 고려)

- 왜 중요?
    - 가장 직관적이고, 거의 모든 트레이더가 씀
    - 단기 타이밍 잡기에 최고

# MACD (추세 + 모멘텀 결합)
- 이동평균 기반 추세 + 모멘텀
- 핵심:
    - MACD선 vs Signal선 골든/데드크로스

- 왜 중요?
    - 추세 전환 포착에 매우 강력
    - RSI보다 “신호 신뢰도”가 높은 편

# ADX (추세 강도 판단)
- 방향이 아니라 **“추세의 힘”**을 측정
- 기준:
    - 20 이하 → 횡보
    - 25 이상 → 추세 있음

- 왜 중요?
    - RSI/MACD 쓰기 전에
    - “지금이 추세장인지?” 판단 가능

# OBV (거래량 기반 확신)
- 가격이 아니라 돈의 흐름
- 핵심:
    - 가격 ↑ + OBV ↑ → 진짜 상승
    - 가격 ↑ + OBV ↓ → 가짜 상승 (다이버전스)
-  왜 중요?
    - 세력(기관) 움직임 추적 가능
    - 가격보다 먼저 움직이는 경우 많음
